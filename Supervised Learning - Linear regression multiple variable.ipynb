<h1> Linear regression with multiple variable # y = m1x1 + m2x2 + m3x3+ .... + c

<h3> Import Libraries

```python
import pandas as pd
import numpy
from sklearn import linear_model
```

```python
#read dataset/file
df = pd.read_csv('multi variable.csv')
```

```python
df
```

```python
exp_fit = df.experience.median() # We will be using the median value to replace the NaN value in experience column
```

```python
exp_fit
```

```python
#replacing NaN values
df.experience = df.experience.fillna(exp_fit)
```

```python
df.experience
```

<h3> creating model

```python
reg = linear_model.LinearRegression() # creating object for linear regression
```

```python
reg.fit(df[['speed', 'car_age', 'experience']], df.risk)
```

<h3> predicting risk value

```python
reg.predict([[160,10,5]])
```

<h3> Finding coefficient

```python
reg.coef_ #value of m1, m2, m3 ....
```

<h3> Intercept 

```python
reg.intercept_ #value of c
```
